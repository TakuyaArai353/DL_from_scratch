```python
# Matplotlib
x = np.arange(0, 6, 0.1)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y2)
```

```python
plt.plot(x, y, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
```
